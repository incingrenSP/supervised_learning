```python
import torch
import torch.nn as nn
import torchvision
from torchvision import transforms, datasets
from torch.utils.data import DataLoader
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
train = datasets.MNIST("", train=True, download=True, transform=transforms.Compose([transforms.ToTensor()]))

test = datasets.MNIST("", train=False, download=True, transform=transforms.Compose([transforms.ToTensor()]))
train_data = DataLoader(train, batch_size=10, shuffle=True)
test_data = DataLoader(test, batch_size=10, shuffle=True)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.dense1 = nn.Linear(784, 64)
        self.dense2 = nn.Linear(64, 64)
        self.dense3 = nn.Linear(64, 64)
        self.dense4 = nn.Linear(64, 10)

    def forward(self, x):
        x = nn.functional.relu(self.dense1(x))
        x = nn.functional.relu(self.dense2(x))
        x = nn.functional.relu(self.dense2(x))
        x = nn.functional.softmax(self.dense2(x), dim=1)

        return x
```

```python
model = NeuralNetwork()
# device = torch.device("cuda:0" if torch.cuda.is_available else "cpu")
# model = model.to(device)
# print(device)

print(model)
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 10

for i in range(epochs):
    for data in train_data:
        X, y = data
        model.zero_grad()
        output = model(X.view(-1, 28*28))
        loss = nn.functional.nll_loss(output, y)
        loss.backward()
        optimizer.step()
    print(loss)
```

```python
correct = 0
total = 0

with torch.no_grad():
    for data in test_data:
        X, y = data
        output = model(X.view(-1, 784))
        for idx, i in enumerate(output):
            if torch.argmax(i) == y[idx]:
                correct += 1
            total += 1

acc = correct / total
print(f'Accuracy: {round(acc, 3)}')
```
